```python
from selenium import webdriver
```

```python
website = 'https://www.adamchoi.co.uk/'

cService = webdriver.ChromeService(executable_path='/Users/apple/Downloads/chromedriver-mac-x64/chromedriver')
driver = webdriver.Chrome(service=cService)
driver.get(website)
```

```python
from selenium.webdriver.common.by import By
one_5 = driver.find_element(By.XPATH,'//table[@id= "desktop-streaks-table"]')
one_5
```

```python
t_row = driver.find_elements(By.XPATH,'//tbody[@class = "ng-scope"]')
len(t_row)
```

```python
import numpy as np
import pandas as pd
tds1 = []
tds2 = []
tds3 = []
tds4 = []
for row in t_row:
    td1 = row.find_element(By.XPATH,'./tr/td[2]')
    if len(td1.text)==0:
        tds1.append(np.nan)
    else:
        tds1.append(td1.text)
    
# td
    td2 = row.find_element(By.XPATH,'./tr/td[3]')
    if len(td2.text)==0:
        tds2.append(np.nan)
    else:
        tds2.append(td2.text)

    td3 = row.find_element(By.XPATH,'./tr/td[4]')
    if len(td3.text)==0:
        tds3.append(np.nan)
    else:
        tds3.append(td3.text)
    
```

```python
dataset = pd.DataFrame({'League': tds1, 'Stat': tds2, 'Next Match': tds3})
dataset
```

```python
dataset.dropna(inplace=True)
dataset
```
